```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
```

```python
for freq in ["143", "217", "353", "545"]:
    table = ascii.read(f"{freq}_source_counts.txt", format="commented_header")
    z_left = table["z_bin_left"]
    fluxes = table["flux_sum"]
    z_bin_width = 0.1
    plt.plot(z_left + z_bin_width/2, fluxes, label=freq)
plt.yscale("log")
plt.legend()
```
